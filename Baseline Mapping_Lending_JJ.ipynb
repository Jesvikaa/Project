### 4.2 Add Area

```python
def area_smeplus_cl(crs):

    if crs['BRANCH'] in ['FATMAWATI','THAMRIN','MT HARYONO']:
        return 'JAKARTA 1A'

    elif crs['BRANCH'] in ['ROXY MAS','ROXY','SUMMITMAS','CEMPAKA MAS']:
        return 'JAKARTA 1B'

    elif crs['BRANCH'] in ['KOTA','KELAPA GADING']:
        return 'JAKARTA 2A'

    elif crs['BRANCH'] in ['EKAJIWA','PLUIT','PURI KENCANA']:
        return 'JAKARTA 2B'

    elif crs['BRANCH'] in ['BANDAR LAMPUNG','BANDAR JAYA','METRO LAMPUNG','BANDAR BENGKULU','TANJUNG KARANG']:
        return 'JAKARTA 2C'

    elif crs['BRANCH'] in ['ALAM SUTRA','SUMMARECON SERPONG','TANGERANG','BOGOR']:
        return 'JAKARTA 3A'

    elif crs['BRANCH'] in ['LIPPO CIKARANG','KARAWANG','BEKASI']:
        return 'JAKARTA 3B'

    elif crs['BRANCH'] in ['KUPANG','DENPASAR','UDAYANA DENPASAR','SUNSET ROAD','MATARAM','JL PEMUDA SURABAYA','MAYJEN SUNGKONO']:
        return 'JATIM A'

    elif crs['BRANCH'] in ['JEMUR ANDAYANI','MALANG','JEMBER','KEDIRI','GRESIK']:
        return 'JATIM B'

    elif crs['BRANCH'] in ['RE MARTADINATA','CIREBON','TASIKMALAYA','GARUT']:
        return 'JAWA BARAT A'

    elif crs['BRANCH'] in ['ASIA AFRIKA','SUKABUMI','KOPO PLAZA']:
        return 'JAWA BARAT B'

    elif crs['BRANCH'] in ['TEGAL','PEKALONGAN','SEMARANG']:
        return 'JAWA TENGAH 1'

    elif crs['BRANCH'] in ['PURWOKERTO','YOGYAKARTA','MAGELANG','SOLO']:
        return 'JAWA TENGAH 2'

    elif crs['BRANCH'] in ['MAKASAR','PALU','KENDARI']:
        return 'SULAWESI'

    elif crs['BRANCH'] in ['MANADO','GORONTALO','JAYAPURA','SORONG','TERNATE','AMBON','MAKASAR']:
        return 'INDONESIA BAG. TIMUR'

    elif crs['BRANCH'] == 'MEDAN':
        return 'SUMATERA UTARA 1'

    elif crs['BRANCH'] in ['PALANG MERAH','RANTAU PRAPAT','KCP THAMRIN']:
        return 'SUMATERA UTARA 2'

    elif crs['BRANCH'] in ['BATAM','PEKANBARU','PADANG']:
        return 'SUMATERA UTARA 3'

    elif crs['BRANCH'] in ['PALEMBANG','JAMBI','MUARA BUNGO']:
        return 'SUMATERA SELATAN'

    elif crs['BRANCH'] in ['PONTIANAK','SAMARINDA']:
        return 'KALIMANTAN A'

    elif crs['BRANCH'] in ['BANJARMASIN','SAMPIT','BALIKPAPAN','SAMARINDA','TARAKAN']:
        return 'KALIMANTAN B'

crs_smeplus_clean['AREA'] = crs_smeplus_clean.apply(area_smeplus_cl, axis=1)
```
